# Oscilador Armónico: Movimiento Armónico Simple

El movimiento armónico simple describe sistemas que oscilan periódicamente:

$x(t) = A\cos(\omega t + \varphi)$

donde $A$ es la amplitud, $\omega$ la frecuencia angular y $\varphi$ la fase inicial.
Modificando estos parámetros podemos explorar distintos comportamientos oscilatorios.

```python
import numpy as np
import matplotlib.pyplot as plt

A = 2.0          # Amplitud (m)
omega = 2*np.pi  # Frecuencia angular (rad/s)  1 Hz
phi = np.pi/4    # Fase inicial (rad)

t = np.linspace(0, 4, 1000)
x = A * np.cos(omega * t + phi)

plt.figure(figsize=(10, 4))
plt.plot(t, x, linewidth=2, color='darkorange')
plt.axhline(0, color='gray', linewidth=0.5)
plt.xlabel('Tiempo (s)')
plt.ylabel('Desplazamiento x (m)')
plt.title(f'Oscilador Armonico  A={A}, ={omega:.2f}, ={phi:.2f}')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## Efecto de los parámetros

Puedes experimentar cambiando los valores de `A`, `omega` y `phi` en la celda anterior:

- **Amplitud (A):** controla la altura máxima de la oscilación.
- **Frecuencia angular (ω):** controla la velocidad de oscilación ($T = 2\pi/\omega$).
- **Fase inicial (φ):** desplaza la curva horizontalmente.

```python
A = 2.0
phi = 0
t = np.linspace(0, 4, 500)

fig, axes = plt.subplots(1, 3, figsize=(14, 3.5), sharey=True)

for ax, omega_val, color in zip(axes, [np.pi, 2*np.pi, 4*np.pi],
                                  ['steelblue', 'darkorange', 'crimson']):
    x = A * np.cos(omega_val * t + phi)
    ax.plot(t, x, linewidth=2, color=color)
    ax.set_title(f' = {omega_val/np.pi:.0f} rad/s')
    ax.set_xlabel('Tiempo (s)')
    ax.grid(True, alpha=0.3)

axes[0].set_ylabel('x (m)')
plt.suptitle('Comparacion de frecuencias', y=1.02)
plt.tight_layout()
plt.show()
```

## Conclusión

El oscilador armónico simple es el modelo básico de fenómenos oscilatorios.
Sistemas reales como péndulos (para ángulos pequeños), resortes y circuitos LC
siguen este comportamiento. La frecuencia angular $\omega = \sqrt{k/m}$ conecta
el movimiento con las propiedades físicas del sistema.
